```python
# Impute most frequent for categorical feature
import pandas as pd
import numpy as np
X = pd.DataFrame({'Shape':['square', 'square', 'oval', 'circle', np.nan]})
X
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
imputer.fit_transform(X)
```

```python
# Impute as the constant 'missing'
imputer = SimpleImputer(strategy='constant', fill_value='missing')
imputer.fit_transform(X)
```

```python
# Multivariate example (Titanic subset)
import pandas as pd
df = pd.read_csv('/content/Titanic-Dataset.csv', nrows=6)
cols = ['SibSp', 'Fare', 'Age']
X = df[cols]
X
```

```python
# IterativeImputer (multivariate imputation)
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
impute_it = IterativeImputer()
impute_it.fit_transform(X)
```

```python
# SimpleImputer univariate mean example
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit([[1, 2], [np.nan, 3], [7, 6]])

# transform example
X = [[np.nan, 2], [6, np.nan], [7, 6]]
print(imp.transform(X))
```

```python
# KNNImputer example
from sklearn.impute import KNNImputer
impute_knn = KNNImputer(n_neighbors=2)
impute_knn.fit_transform(X)
```

```python
# Using missingness as a feature (indicator)
import pandas as pd
import numpy as np
X = pd.DataFrame({'Age':[20, 30, 10, np.nan, 10]})
X

from sklearn.impute import SimpleImputer
# impute the mean
imputer = SimpleImputer()
imputer.fit_transform(X)

# with indicator
imputer = SimpleImputer(add_indicator=True)
imputer.fit_transform(X)
```

```python
df = pd.read_csv("/content/Titanic-Dataset.csv")
# select top two rows where column 'Sex' is 'female'
df [df ['Sex'] == 'female'].head(2)
# multiple conditions (female AND age >= 55)
df [(df ['Sex'] == 'female') & (df ['Age'] >= 55)]
```

```python
# replace any instance of 'female' with 'Woman' in the 'Sex' column
df ['Sex'].replace('female', 'Woman').head(2)

# replace multiple instances ('female' with 'Woman', 'male' with 'Man')
df ['Sex'].replace(['female', 'male'], ['Woman', 'Man']).head(5)

# replace all instances of value 1 with string "One" across the DataFrame
df.replace(1, "One").head(2)
```

```python
# rename a single column
df.rename(columns={'PClass': 'Passenger Class'}).head(2)

# rename multiple columns
df.rename(columns={'PClass': 'Passenger Class', 'Sex': 'Gender'}).head(2)
```

```python
print('Maximum: {}'.format(df['Age'].max()))
print('Minimum: {}'.format(df ['Age'].min()))
print('Mean: {}'.format(df ['Age'].mean()))
print('Sum: {}'.format(df['Age'].sum()))
print('Count: {}'.format(df['Age'].count()))
```

```python
print("Variance:", df['Fare'].var())
print("Standard Deviation:", df['Fare'].std())
print("Kurtosis:", df['Fare'].kurt())
print("Skewness:", df['Fare'].skew())
```

```python
# unique will return an array of all unique values in a column
df ['Sex'].unique()

# value_counts will display all unique values with the number of times each value appears
df ['Sex'].value_counts()
```

```python
# select missing values, show 2 rows
df [df ['Age'].isnull()].head(2)
```

```python
# axis=1 means the column axis
df.drop('Age', axis=1).head(2)
```

```python
# create new dataframe excluding the rows where 'Sex' is 'male'
df [df ['Sex'] != 'male'].head(2)

# delete a row by matching a unique value ('Name')
df [df ['Name'] != 'Allison, Miss Helen Loraine'].head(2)
# delete a row by index (index 0)
df [df.index != 0].head(2)
```

```python
# group by 'Sex' and calculate the mean of all numerical columns
df.groupby('Sex')[df.select_dtypes(include='number').columns].mean()

# group by 'Survived' and count the number of 'Name' entries
df.groupby('Survived') ['Name'].count()

# group by multiple columns ('Sex' and 'Survived') and calculate the mean of 'Age'
df.groupby(['Sex', 'Survived']) ['Age'].mean()
```

```python
# for loop
for name in df ['Name'] [0:2]:
    print(name.upper())

# list comprehension (more "pythonic")
[name.upper() for name in df ['Name'] [0:2]]
```

```python
# group by 'Sex' and apply the count function to all columns in each group
df.groupby('Sex').apply(lambda x: x.count())
```
